```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Top 10 Medicine Sub-Categories
plt.figure(figsize=(10, 6))
sns.countplot(y='sub_category', data=df, order=df['sub_category'].value_counts().head(10).index)
plt.title("Top 10 Medicine Sub-Categories")
plt.xlabel("Count")
plt.ylabel("Sub Category")
plt.show()
```

```python
# Top 10 Most Common Medicines
plt.figure(figsize=(10, 6))
sns.countplot(y='product_name', data=df, order=df['product_name'].value_counts().head(10).index)
plt.title("Top 10 Most Frequent Products")
plt.xlabel("Count")
plt.ylabel("Product Name")
plt.show()
```

```python
# Top 5 Manufacturers
top_manufacturers = df['product_manufactured'].value_counts().head(5)
plt.figure(figsize=(6, 6))
plt.pie(top_manufacturers, labels=top_manufacturers.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Medicine Manufacturers")
plt.axis('equal')
plt.show()
```

```python
# Top 5 Side Effects
top_effects = df['side_effects'].value_counts().head(5)
plt.figure(figsize=(6, 6))
plt.pie(top_effects, labels=top_effects.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 10 Reported Side Effects")
plt.axis('equal')
plt.show()
```

```python
# Medicine Price Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['product_price'], bins=10, kde=True)
plt.title("Distribution of Medicine Prices")
plt.xlabel("Price")
plt.ylabel("Number of Products")
plt.show()
```
