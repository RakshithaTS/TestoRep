```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Study Status', palette='Set2')
plt.title("Clinical Trial Status for Cancer")
plt.ylabel("Number of Trails")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Study Type', hue='Study Status', palette='Set2')
plt.title("Study Type vs. Study Status")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
top_conditions = df['Conditions'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_conditions.index, x=top_conditions.values, palette='Set3')
plt.title("Top 10 Conditions in Cnacer Clinical Trails")
plt.xlabel("Number of Trails")
plt.ylabel("Condition")
plt.tight_layout()
plt.show()
```
